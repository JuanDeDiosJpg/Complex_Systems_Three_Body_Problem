\end{equation}

\begin{equation}
    \tilde{P_{r,i}}=\frac{P_{r,i}}{m_id_{TL}}

\begin{equation}

```python
import sympy as sp
from sympy import symbols, init_printing
from IPython.display import display, Math
import numpy as np

H=sp.Symbol('H')

Ache=sp.Symbol('H')

G=sp.Symbol('G')

dTL=sp.Symbol('d_{TL}')

r1=sp.Symbol('r_1')
r2=sp.Symbol('r_2')
r3=sp.Symbol('r_3')

Pr1=sp.Symbol('P_{r1}')
Pr2=sp.Symbol('P_{r2}')
Pr3=sp.Symbol('P_{r3}')

#phi1=sp.Symbol('\phi_1')
#phi2=sp.Symbol('\phi_2')
#phi3=sp.Symbol('\phi_3')

#Pphi1=sp.Symbol('Pphi1')
#Pphi2=sp.Symbol('Pphi2')
#Pphi3=sp.Symbol('Pphi3')

phi1, phi2, phi3, Pphi1, Pphi2, Pphi3=symbols('phi_1 phi_2 phi_3 P_{phi_1} P_{phi_2} P_{phi_3}')

phi1punto, phi2punto, phi3punto, Pphi1punto, Pphi2punto, Pphi3punto=symbols(r'\dot{\phi}_{1} \dot{\phi}_{2} \dot{\phi}_{3} \dot{P_\phi}_{1} \dot{P_\phi}_{2} \dot{P_\phi}_{3}')
r1punto, r2punto, r3punto, Pr1punto, Pr2punto, Pr3punto=symbols(r'\dot{r}_{1} \dot{r}_{2} \dot{r}_{3} \dot{P_r}_{1} \dot{P_r}_{2} \dot{P_r}_{3}')

#Pongo todas las variables derivadas tildadas aquí ahora 
r1tildepunto, r2tildepunto, r3tildepunto, Pr1tildepunto, Pr2tildepunto, Pr3tildepunto=symbols(r'\dot{\tilde{r}}_{1} \dot{\tilde{r}}_{2} \dot{\tilde{r}}_{3} \dot{\tilde{P_r}}_{1} \dot{\tilde{P_r}}_{2} \dot{\tilde{P_r}}_{3}')
phi1tildepunto, phi2tildepunto, phi3tildepunto, Pphi1tildepunto, Pphi2tildepunto, Pphi3tildepunto=symbols(r'\dot{\tilde{\phi}}_{1} \dot{\tilde{\phi}}_{2} \dot{\tilde{\phi}}_{3} \dot{\tilde{P_\phi}}_{1} \dot{\tilde{P_\phi}}_{2} \dot{\tilde{P_\phi}}_{3}')

#Pongo todas las variables tildadas normales aquí
#Pongo todas las variables derivadas tildadas aquí ahora 
r1tilde, r2tilde, r3tilde, Pr1tilde, Pr2tilde, Pr3tilde=symbols(r'\tilde{r}_{1} \tilde{r}_{2} \tilde{r}_{3} \tilde{P_r}_{1} \tilde{P_r}_{2} \tilde{P_r}_{3}')
phi1tilde, phi2tilde, phi3tilde, Pphi1tilde, Pphi2tilde, Pphi3tilde=symbols(r'\tilde{\phi}_{1} \tilde{\phi}_{2} \tilde{\phi}_{3} \tilde{P_\phi}_{1} \tilde{P_\phi}_{2} \tilde{P_\phi}_{3}')

#Doy las variables sustituidas por las tildadas
r1escal=r1tilde*dTL
r2escal=r2tilde*dTL
r3escal=r3tilde*dTL

Pr1escal=Pr1tilde*m1*dTL
Pr2escal=Pr2tilde*m2*dTL
Pr3escal=Pr3tilde*m3*dTL

Pphi1escal=Pphi1tilde*m1*dTL**2
Pphi2escal=Pphi2tilde*m2*dTL**2
Pphi3escal=Pphi3tilde*m3*dTL**2

r1puntoescal=r1tildepunto*dTL
r2puntoescal=r2tildepunto*dTL
r3puntoescal=r3tildepunto*dTL

Pr1puntoescal=Pr1tildepunto*m1*dTL
Pr2puntoescal=Pr2tildepunto*m2*dTL
Pr3puntoescal=Pr3tildepunto*m3*dTL

Pphi1puntoescal=Pphi1tildepunto*m1*dTL**2
Pphi2puntoescal=Pphi2tildepunto*m2*dTL**2
Pphi3puntoescal=Pphi3tildepunto*m3*dTL**2


m1=sp.Symbol('m_1')
m2=sp.Symbol('m_2')
m3=sp.Symbol('m_3')
#Creo el Hamiltoniano ahora


Hcin=m1/2*((Pr1**2/m1**2)+Pphi1**2/(m1**2*r1**2))+m2/2*((Pr2**2/m2**2)+Pphi2**2/(m2**2*r2**2))+m3/2*((Pr3**2/m3**2)+Pphi3**2/(m3**2*r3**2))

Hpot1=-G*m1*(m2/(r2**2+r1**2-2*r2*r1*sp.cos(phi2-phi1))**(1/2)+m3/(r3**2+r1**2-2*r3*r1*sp.cos(phi3-phi1))**(1/2))

Hpot2=-G*m2*(m3/(r3**2+r2**2-2*r3*r2*sp.cos(phi3-phi2))**(1.0/2.0))



H=Hcin+Hpot1+Hpot2
ecuacion0=sp.Eq(Ache, H)

#Ahora hago las derivadas
dH_dr1=sp.diff(H, r1)
ecuacion1=sp.Eq(Pr1tildepunto, (1/(m1*dTL))*sp.simplify(-dH_dr1.subs({r1: r1escal, r2: r2escal, r3: r3escal, Pphi1:Pphi1escal})))

dH_dr2=sp.diff(H, r2)
ecuacion2=sp.Eq(Pr2tildepunto, (1/(m2*dTL))*sp.simplify(-dH_dr2.subs({r1: r1escal, r2: r2escal, r3: r3escal, Pphi2:Pphi2escal})))

dH_dr3=sp.diff(H, r3)
ecuacion3=sp.Eq(Pr3tildepunto, (1/(m3*dTL))*sp.simplify(-dH_dr3.subs({r1: r1escal, r2: r2escal, r3: r3escal, Pphi3:Pphi3escal})))

dH_dPr1=sp.diff(H, Pr1)
ecuacion4=sp.Eq(r1tildepunto, 1/dTL*dH_dPr1.subs({Pr1: Pr1escal}))

dH_dPr2=sp.diff(H, Pr2)
ecuacion5=sp.Eq(r2punto, 1/dTL*dH_dPr2.subs({Pr2: Pr2escal}))

dH_dPr3=sp.diff(H, Pr3)
ecuacion6=sp.Eq(r3punto, 1/dTL*dH_dPr3.subs({Pr3: Pr3escal}))


dH_dphi1=sp.diff(H, phi1)
ecuacion7=sp.Eq(Pphi1tildepunto, 1/(m1*dTL**2)*sp.simplify(-dH_dphi1.subs({r1: r1escal, r2: r2escal, r3: r3escal})))

dH_dphi2=sp.diff(H, phi2)
ecuacion8=sp.Eq(Pphi2tildepunto, 1/(m2*dTL**2)*sp.simplify(-dH_dphi2.subs({r1: r1escal, r2: r2escal, r3: r3escal})))

dH_dphi3=sp.diff(H, phi3)
ecuacion9=sp.Eq(Pphi3tildepunto, 1/(m3*dTL**2)*sp.simplify(-dH_dphi3.subs({r1: r1escal, r2: r2escal, r3: r3escal})))

dH_dPphi1=sp.diff(H, Pphi1)
ecuacion10=sp.Eq(phi1punto, dH_dPphi1.subs({Pphi1: Pphi1escal, r1: r1escal}))

dH_dPphi2=sp.diff(H, Pphi2)
ecuacion11=sp.Eq(phi2punto, dH_dPphi2.subs({Pphi2: Pphi2escal, r2: r2escal}))

dH_dPphi3=sp.diff(H, Pphi3)
ecuacion12=sp.Eq(phi3punto, dH_dPphi3.subs({Pphi3: Pphi3escal, r3: r3escal}))



# Inicializamos la impresión bonita
init_printing()

print("Calculo el Hamiltoniano")
display(Math(sp.latex(ecuacion0)))

print("\nCalculo Pr1 punto (la menos derivada de H respecto a r1)")
display(Math(sp.latex(ecuacion1)))

print("\nCalculo  Pr2 punto")
display(Math(sp.latex(ecuacion2)))

print("\nCalculo  Pr3 punto")
display(Math(sp.latex(ecuacion3)))

print("\ncalcul r1 punto")
display(Math(sp.latex(ecuacion4)))

print("\ncalculó r2 punto")
display(Math(sp.latex(ecuacion5)))

print("\ncalculó r3 punto")
display(Math(sp.latex(ecuacion6)))

print('\nCalcuo Pphi1 punto')
display(Math(sp.latex(ecuacion7)))

print('\nCalcuo Pphi2 punto')
display(Math(sp.latex(ecuacion8)))

print('\nCalcuo Pphi3 punto')
display(Math(sp.latex(ecuacion9)))


print('\nCalculo phi1 punto')
display(Math(sp.latex(ecuacion10)))

print('\nCalculo phi2 punto')
display(Math(sp.latex(ecuacion11)))

print('\nCalculo phi3 punto')
display(Math(sp.latex(ecuacion12)))


#Quiero que me dé todo en latex para poder guardarlo en la presentación

print(sp.latex(ecuacion0))

print(sp.latex(ecuacion1))
print(sp.latex(ecuacion2))
print(sp.latex(ecuacion3))
print(sp.latex(ecuacion4))
print(sp.latex(ecuacion5))
print(sp.latex(ecuacion6))
print(sp.latex(ecuacion7))
print(sp.latex(ecuacion8))
print(sp.latex(ecuacion9))
print(sp.latex(ecuacion10))
print(sp.latex(ecuacion11))
print(sp.latex(ecuacion12))
```

# ANIMACIONES

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation

M0=np.loadtxt("Masa0.txt")
M1=np.loadtxt("Masa1.txt")
M2=np.loadtxt("Masa2.txt")

x0=M0[0:500000:100, 0]
x1=M1[0:500000:100, 0]
x2=M2[0:500000:100, 0]

y0=M0[0:500000:100,1]
y1=M1[0:500000:100,1]
y2=M2[0:500000:100,1]

fig, ax=plt.subplots()

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Sun, Earth and Mars orbit")
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)


line1, =ax.plot(x0, y0, label=r'$M_1$')
line2, =ax.plot(x1, y1, label=r'$M_2$')
line3, =ax.plot(x2, y2, label=r'$M_3$')

#Esto es para que se vean los planetas por encima de la trayectoria que dejan
point1, = ax.plot([], [], 'o', c=line1.get_color())
point2, = ax.plot([], [], 'o', c=line2.get_color())
point3, = ax.plot([], [], 'o', c=line3.get_color())

ax.legend()

def anim(frame):
    line1.set_data(x0[:frame], y0[:frame])
    line2.set_data(x1[:frame], y1[:frame])
    line3.set_data(x2[:frame], y2[:frame])


    #Mostramos los planetas por encima de la trayectoria
    if frame > 0:
        point1.set_data([x0[frame-1]], [y0[frame-1]])
        point2.set_data([x1[frame-1]], [y1[frame-1]])
        point3.set_data([x2[frame-1]], [y2[frame-1]])

    return line1, line2, line3, point1, point2, point3
    #return point1, point2, point3

plt.show()
ani=animation.FuncAnimation(fig, anim, frames=5000, interval=60, blit=True)
ani.save(filename="Sun_Earth_Mars.gif", writer='pillow')
```
